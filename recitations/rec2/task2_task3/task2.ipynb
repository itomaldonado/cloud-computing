```python
import findspark
findspark.init()
from pyspark.sql import SQLContext
from pyspark import SparkContext
from pyspark.mllib.linalg.distributed import CoordinateMatrix, MatrixEntry, RowMatrix

sc = SparkContext(appName="task2")

sqlContext = SQLContext(sc)
lines = sc.textFile("part-00000")

def A(t):
    if t[0]=='A': return True
    else: return False
    
def B(t):
    if t[0]=='B': return True
    else: return False

####read entries from file and map to a MatrixEntries
entries_A=lines.map(lambda x: x.split(" ")).filter(A).map(lambda word:MatrixEntry(word[1],word[2],word[3]))
entries_B=lines.map(lambda x: x.split(" ")).filter(B).map(lambda word:MatrixEntry(word[1],word[2],word[3]))

####create Coordinate Matrix
mat_A = CoordinateMatrix(entries_A)
mat_B = CoordinateMatrix(entries_B)

mat_A.entries.collect()
mat_B.entries.collect()
```

```python
####convert to row format
rowMat_A = mat_A.toRowMatrix()
rowMat_B = mat_B.toRowMatrix()

rowMat_A.rows.collect()
rowMat_B.rows.collect()
```
